## emohawk demo

```python
import emohawk
```

### Reading GRIB data

```python
data = emohawk.open("data/era5.grib")
data  # Note that emohawk automatically detects that the input file is GRIB
```

```python
# emohawk offers conversion to widely-used file formats and Python objects
[m for m in dir(data) if m.startswith("to_")]
```

```python
data.to_xarray()
```

```python
data.to_numpy()
```

```python
# emohawk also has convenience methods for quickly checking dimensions or axes of data
data.axis("x")
```

```python
# emohawk is also able to read zipped files (currently single zipped files only)
data = emohawk.open("data/efas.zip")
data  # Note that we have a netCDF object, not a zip object
```

```python
data.to_pandas()
```
